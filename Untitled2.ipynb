# Version Control System:

 * A version control system is a software that helps software that helps software developers to work together and maintain a complete history of their work.
 
### Functionalities of VCS:

 * Allow developers to work simultaneously
 * It doesnot allow overwriting of each other changes
 * Maintain a history of every version
 * VCS is a system that records changes to a files/set of files over the time so that we can recall specific version later
 
## Types of VCS:
* centralised version control system
* De-centralised Version Control system
